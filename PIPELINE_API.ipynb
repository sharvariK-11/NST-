```python
from transformers import pipeline
```

### sentiment analysis pipeline


```python
classifier= pipeline("sentiment-analysis")
classifier ("I am fine")
```

```python
classifier = pipeline("sentiment-analysis")
classifier( ["I don't like chocolate", "How are you?", "I like cycling"])
```

```python
classifier=pipeline("zero-shot-classification")

classifier(
    "This is good", 
    candidate_labels=["education", "food"],
    )
```

```python
generator=pipeline("text-generation")
generator("I this course we will teach you how to")
generator("I will")
```

```python
generator= pipeline("text-generation", model="distilgpt2")
generator(
    "I will",
    max_length =30,
    num_return_sequences=2,
)
```

###fill mask pipeline 
fills the mask word by guess word


>



```python
unmasker=pipeline("fill-mask")
unmasker("This will teach you all about <mask> models", top_k=2)
```

### entity (person, locatio) identification

```python
ner= pipeline("ner", grouped_entities=True)
ner("My name is Sharvari, I stay in Mumbai at IIT ")
```

###Question answering


```python
question_answerer= pipeline("question-answering")
question_answerer(
    question="what do I eat?",
    context= "I am Sharvari, I eat apple"
)
```

### Summariser


```python
summarizer=pipeline("summarization")
summarizer("hello, its pleasure to meet you. You are a great person, I adore you.You are a role model to me .I am inspired a lotLuckily, I got my inspiration at my home. She is my mother, mentor, a very good friend, etc. But all in one she is an amazing personality and a role model for all of us. I wish I could become at least a bit like her. I never get disappointed because she is always with me to take care of me and also encourages me to do something different. All of us learn from different people and she is one of them. Apart from her I also want to appreciate my teacher and friends who also play a very important role in motivating me.")
```

```python
from transformers import pipeline
unmasker=pipeline("fill-mask", model="bert-base-uncased")
result=unmasker("This woman works as a [MASK].")
print([r["token_str"]for r in result])

result=unmasker("This man works as a [MASK].")
print([r["token_str"]for r in result])
```
